# 1 Задача - ответить на вопрос есть ли связь между жёсткостью воды и средней годовой смертностью?

```python
# импорт необходимых библиотек

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_profiling as pdp
```

```python
# импорт данных
df = pd.read_csv('water.csv')
```

# Построим точечный график и рассчитаем коэффициенты зависимости Спирмена, Пирсона и Кенделла

```python
#Создание отчета по разведочному анализу данных
report = pdp.ProfileReport(df, title='Разведочный анализ данных')
report
```

```python
# Еще раз построим график зависимости сметрности от жесткости воды, чтобы проверить, что библиотека Pandas-profiling не врет (а она не врет)
from pylab import rcParams
rcParams['figure.figsize'] = 16,9

df.plot(kind='scatter', x='hardness', y='mortality')
```

```python
#Рассчитаем значения коэф-та Пирсона
df[['hardness', 'mortality']].corr()
```

```python
#Рассчитаем значения коэф-та Спирмена
df[['hardness', 'mortality']].corr(method='spearman')
```

```python
#Рассчитаем значения коэф-та Кенделла
df[['hardness', 'mortality']].corr(method='kendall')
```

# Выводы по итогам разведочного анализа данных

Бибилотека pandas-profiling уже построила визуализации всех трех коэффициентов корреляции (Пирсона, Спримена и Кенделла).
Две из трех визуализаций (Спирмена и Пирсона) показывают существование сильной обратной взаимосвязи между смертностью и жесткостью воды в Великобритании. Визуализация Кенделла показывает существование средней обратной связи между параметрами.

Поскольку, исходя из графика зависимости сметрности от жесткости воды, нельзя однозначно сказать, что зависимость линейная, то будем придерживаться значения коэффициента Кенделла, который чуть ниже, чем остальные (-0,454).

# Построим модель линейной регрессии

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Обозначим нашу функцию и аргументы
X = df[['hardness']]
y = df['mortality']
```

```python
#Разобьем набор данных на учебную часть и тестовую
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
 # обучаем модель
model = LinearRegression()
model.fit(X_train, y_train) 
```

```python
model.coef_
```

```python
model.intercept_
```

Функция линейной регрессии получилась следующая:
у = -3.13167087x + 1668.9722977277095

```python
# Получим пронозные значения y на тестовых данных:
y_pred = model.predict(X_test)
y_pred
```

```python
# Строим график линейной регрессии на основе тестового X и прогнозного Y:
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, c='r')
```

# Определим коэффициент детерминации

```python
model.score(X_test, y_test)
```

Получаем среднюю точность предсказательной модели, что видно из графика, в принципе.

# Анализ остатков

```python
residuals = y_pred - y_test
sns.regplot(x = residuals, y = y_pred, data = None, scatter = True, color = 'red')
plt.title('Residuals')
```

Точки случайно распределены с обеих сторон нашей прямой, значит, в данном случае модель линейной регрессии в принципе подходит для предсказания зависимости смертности от уровня жесткости воды

# 2 Задание. Сохраняется ли аналогичная зависимость для северных и южных городов по отдельности?

```python
# Создадим 2 датафрейма отдельно для северных, отдельно для южных городов

df_north = df[df.location =='North']
df_south = df[df.location =='South']
```

```python
#Построим график зависимости сметрности от жесткости воды для северных городов
df_north.plot(kind='scatter', x='hardness', y='mortality')
```

```python
#Построим график зависимости сметрности от жесткости воды для северных городов
df_south.plot(kind='scatter', x='hardness', y='mortality')
```

```python
#Рассчитаем значения коэф-та Пирсона для северных городов
df_north[['hardness', 'mortality']].corr()
```

```python
#Рассчитаем значения коэф-та Пирсона для южных городов
df_south[['hardness', 'mortality']].corr()
```

```python
#Рассчитаем значения коэф-та Спирмена  для северных городов
df_north[['hardness', 'mortality']].corr(method='spearman')
```

```python
#Рассчитаем значения коэф-та Спирмена  для южных городов
df_south[['hardness', 'mortality']].corr(method='spearman')
```

```python
#Рассчитаем значения коэф-та Кенделла для северных городов
df_north[['hardness', 'mortality']].corr(method='kendall')
```

```python
#Рассчитаем значения коэф-та Кенделла для южных городов
df_south[['hardness', 'mortality']].corr(method='kendall')
```

# Выводы по итогам разведочного анализа данных для северных и южных городов по отдельности
По аналогии с общей выборкой будем придерживаться коэффициента Кенделла. Как видно, коэффициент Кенделла равен -0,283 для северной части и -0,44 для южной части Великобритании, что характерно для средней обратной связи между параметрами. Однако сила взаимосвязи для северных городов меньше, чем для южных.

По итогам анализа трех наборов данных можно с уверенностью говорить о существовании статистически значимой обратной корреляционной связи между степенью жесткости питьевой воды и уровнем смертности населения как для страны в целом, так и в отдельности для северных и южных городов. Однако, для южных городов данная взаимосвязь проявляется сильнее.

# Построим модель линейной регрессии для северных городов

```python
Xn = df_north[['hardness']]
yn = df_north['mortality']

Xn_train, Xn_test, yn_train, yn_test = train_test_split(Xn, yn, test_size=0.30, random_state=42)

model_north = LinearRegression()
model_north.fit(Xn_train, yn_train)

yn_pred = model_north.predict(Xn_test)

plt.scatter(Xn_test, yn_test)
plt.plot(Xn_test, yn_pred, c='r')
```

```python
model_north.coef_
```

```python
model_north.intercept_
```

```python
model_north.score(Xn_test, yn_test)
```

Линейная регрессия для северных городов принимает вид
у = -1.97686226x + 1688.150229031281

Коэффициент детерминации = 0.019, что является довольно низким значением, а значит, полученная модель обладает очень низкой точностью.

Вероятно, это связано с тем, что у нас было мало исходных данных для обучения модели.

# Анализ остатков для северных городов

```python
residuals_north = yn_pred - yn_test
sns.regplot(x = residuals_north, y = yn_pred, data = None, scatter = True, color = 'red')
plt.title('Residuals for north towns')
```

Точки случайно распределены с обеих сторон нашей прямой, значит, в данном случае модель линейной регрессии, в принципе, подходит для предсказания зависимости смертности от уровня жесткости воды для северных городов.

# Построим модель линейной регрессии для южных городов

```python
Xs = df_south[['hardness']]
ys = df_south['mortality']

Xs_train, Xs_test, ys_train, ys_test = train_test_split(Xs, ys, test_size=0.30, random_state=42)

model_south = LinearRegression()
model_south.fit(Xs_train, ys_train)

ys_pred = model_south.predict(Xs_test)

plt.scatter(Xs_test, ys_test)
plt.plot(Xs_test, ys_pred, c='r')
```

```python
model_south.coef_
```

```python
model_south.intercept_
```

```python
model_south.score(Xs_test, ys_test)
```

Линейная регрессия для южных городов принимает вид у = -3.51342921x + 1640.2204603423374

Коэффициент детерминации = -2,55, что означает, что полученная линейная модель не подходит для описания корреляционной связи

Вероятно, это связано с тем, что у нас было мало исходных данных для обучения модели, как в случае с северными городами.

# Анализ остатков для южных городов

```python
residuals_south = ys_pred - ys_test
sns.regplot(x = residuals_south, y = ys_pred, data = None, scatter = True, color = 'red')
plt.title('Residuals for south towns')
```

Точки случайно распределены с обеих сторон нашей прямой, значит, в данном случае модель линейной регрессии, в принципе, подходит для предсказания зависимости смертности от уровня жесткости воды для южных городов.
