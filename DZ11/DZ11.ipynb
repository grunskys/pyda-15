```python
#Задание 1
#Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
#- оценка 2 и меньше - низкий рейтинг
#- оценка 4 и меньше - средний рейтинг
#- оценка 4.5 и 5 - 'высокий рейтинг'

import pandas as pd

# взят датасет ml-latest-small 
 
rating = pd.read_csv('ratings.csv')

#Присвоим оценкам переменные для универсальности
extra_high = 5.0
high = 4.5
middle = 4
low = 2

#Получаем Dataframe со средним рейтингом, сгруппированный по ID фильма
df = rating[['movieId','rating']].groupby('movieId').mean().reset_index()

#Функция присвоения рейтинга
def avg_rating(row):
    if row['rating'] <= low:
        return 'низкий рейтинг'
    elif row['rating'] <= middle:
        return 'средний рейтинг'
    elif row['rating'] == high or row['rating'] == extra_high:
        return 'высокий рейтинг'
    else:
        return 'странный рейтинг'
        
    
df['avg_rating'] = df.apply(avg_rating, axis=1)
df.head(5010)
```

```python
#Задание 2. Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

import re

region = pd.read_csv('keywords.csv')

geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}

#Функция поиска значений словаря в запросах
def region_find(row):
    for key, value in geo_data.items(): #Начинаем перебор по ключам, значениям
            for word in value: #Распаковываем значения, перебрая его отдельные элементы
                # Поскольку в поисковых запросах города могут быть использованы в разных падежах, то прибегнем к использованию регулярных выражений
                #для этого, сначала надо отсечь окончания, где они есть. если где-то нет окончаний, то это не проблема, вхожление все равно будет
                length = len(word)-1 #высчитываем, сколько букв от названия города оставим для будущего шаблона поисковой регулярки
                pattern = r'^\w{5}' #пишем шаблон регулярки. цифру можно использовать любую, вместо нее будет подставлена переменная length
                pattern_sub = re.sub(r'\d', str(length), pattern) #заменяем цифру в шаблоне на значение переменной length
                city = re.match(pattern_sub, word) #получаем город с "отрезанным" окончанием, который потом будем сравнивать со значением в keywords
                matched = re.search(city[0], row['keyword'].lower()) # ищем конкретное вхождение в значении поискового запроса
                if matched:
                    return key
                else:
                    return 'undefined'

region['region'] = region.apply(region_find, axis=1)
region.head(200)
    
```
